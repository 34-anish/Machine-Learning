```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df[['price']])
reg.predict([[4000]])
```
