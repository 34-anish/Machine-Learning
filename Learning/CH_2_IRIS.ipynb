```python
import pandas  as pd
from sklearn.datasets import load_iris
```

#DATA

```python
iris = load_iris()
```

```python
iris
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
```

```python
df.tail()
```

```python
df['flower'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df
```

```python
df['flower'].value_counts() 
```

```python
df0 = df.iloc[df['flower'].values  =='setosa']
df1 = df.iloc[df['flower'].values  =='versicolor']
df2 = df.iloc[df['flower'].values  =='virginica']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color ='green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color ='blue',marker ='x')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color ='red',marker ='<')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.plot()
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color ='green')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color ='blue',marker ='x')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color ='red',marker ='<')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.plot()
```
