```python
import numpy as np
import matplotlib.pyplot as plt
```

Step Input Unit Function
 $1- e^(\frac{-t}{\tau} )$

```python
x = np.linspace(0,13,100)
y1 = 1-np.exp(-x/1)
y2 = 1-np.exp(-x/2)
y3 = 1-np.exp(-x/3)

plt.figure(figsize=(10,10))
plt.plot(x, y1, '-r', label='tau = 1')
plt.plot(x, y2, '-g', label='tau = 2')
plt.plot(x, y3, '-b', label='tau = 3')


plt.title(r'Step input test signal $1- e ^ (\frac{-t}{\tau}) $ ', fontsize='small')
plt.xlabel('t')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

Step Input Unit Function
$t - \tau + \tau * e^(\frac{-t}{\tau}) $ 

```python

t = np.linspace(0,13,100)
y1 = t - 1 + np.exp(-t)
y2 = t - 2 + 2*np.exp(-t/2)
y3 = t - 3 + 3*np.exp(-t/3)

plt.figure(figsize=(10,10))
plt.plot(x, y1, '-r', label='tau = 1')
plt.plot(x, y2, '-g', label='tau = 2')
plt.plot(x, y3, '-b', label='tau = 3')


plt.title(r'Unit Ramp $t - \tau + \tau * e^(\frac{-t}{\tau}) $ ', fontsize='small')
plt.xlabel('t')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

Unit Impulse
$\frac{1}{\tau} * e^(\frac{-t}{\tau}) $ 

```python


t = np.linspace(0,20,100)
y1 = 1 *np.exp(-t)
y2 = 1/2 * np.exp(-t/2)
y3 = 1/3 * np.exp(-t/3)

plt.figure(figsize=(10,10))
plt.plot(x, y1, '-r', label='tau = 1')
plt.plot(x, y2, '-g', label='tau = 2')
plt.plot(x, y3, '-b', label='tau = 3')


plt.title(r'Unit Impulse $\frac{1}{\tau} * e^(\frac{-t}{\tau}) $ ', fontsize='small')
plt.xlabel('t')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
