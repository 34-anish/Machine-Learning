# Loss And Cost Function
Individual error is loss and the cumulative error is cost function


```python
import numpy as np
```

```python
y_predicted = [1,0,1,1,0]
y_true =[0.9,0.1,0.5,0.6,0]
```

```python
def mae(y_predicted , y_true):
    error = 0
    for yp,yt in zip(y_predicted ,y_true):
        error += abs(yp-yt)
    return error/len(y_predicted)
    
```

```python
mae(y_predicted,y_true)
```

# Mean Squared Error

```python
def mse(y_predicted , y_true):
    error = 0
    for yp,yt in zip(y_predicted ,y_true):
        error += (yp-yt)**2
    return error/len(y_predicted)
    
```

```python
mse(y_predicted,y_true)
```

```python
np.mean(np.square(y_predicted-y_true))
```
