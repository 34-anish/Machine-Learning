# Sigmoid Functions


where $$ z =  wx+b$$

```python
import math
def sigmoid(x) :
    return 1/(1+math.exp(-x))
```

```python
sigmoid(1)
```

### Tanh


```python
def tanh(z):
    return (math.exp(z)-math.exp(-z))/(math.exp(z)+math.exp(-z))
```

```python
tanh(1)
```

## Issues with sigmoid and tanh 
- It has vanishing gradient

```python
def relu(x):
    return max(0,x)
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
relu(-100),leaky_relu(-100)
```

Relu can be used in hidden layers(default choice if we are unsure about the activation function)
